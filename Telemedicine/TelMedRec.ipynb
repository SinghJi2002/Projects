```python
import pandas as pd
import numpy as np
import torch
```

```python
#Loading Dataset.
dataframe=pd.read_excel("TeleMedicine.xlsx")
#Converting integer type columns to string for searching
dataframe['Age']=dataframe["Age"].astype(str)
dataframe['PatientID']=dataframe["PatientID"].astype(str)
dataframe=dataframe.set_index("PatientID")
#Converting each element to uppercase for convienence in searching
dataframe["Medical Condition"] = dataframe["Medical Condition"].apply(lambda x: x.upper())
dataframe["Other Demographics"] = dataframe["Other Demographics"].apply(lambda x: x.upper())
dataframe["Insurance Type"] = dataframe["Insurance Type"].apply(lambda x: x.upper())
dataframe["Tech Device"] = dataframe["Tech Device"].apply(lambda x: x.upper())
dataframe["Tech Access"] = dataframe["Tech Access"].apply(lambda x: x.upper())
dataframe["Symptoms"] = dataframe["Symptoms"].apply(lambda x: x.upper())
```

```python
for i in range(0,len(dataframe),1):
  if(dataframe['Medical Condition'].iloc[i]=='CHRONIC ARTHRITIS'):
    dataframe['Medication'].iloc[i]="NSAIDs, DMARDs, corticosteroids, biologic agents"
    dataframe['Treatment'].iloc[i]="Physical therapy, occupational therapy, joint injections, surgery"
  elif(dataframe['Medical Condition'].iloc[i]=='ASTHMA'):
    dataframe['Medication'].iloc[i]="Bronchodilators, inhaled corticosteroids"
    dataframe['Treatment'].iloc[i]="Inhaler techniques, peak flow monitoring, allergy management"
  elif(dataframe['Medical Condition'].iloc[i]=='HIGH CHOLESTEROL'):
    dataframe['Medication'].iloc[i]="Statins, bile acid sequestrants, fibrates"
    dataframe['Treatment'].iloc[i]="Lifestyle modifications, lipid-lowering therapy, regular monitoring"
  elif(dataframe['Medical Condition'].iloc[i]=='CHRONIC SKIN CONDITION'):
    dataframe['Medication'].iloc[i]="Topical corticosteroids, oral medications"
    dataframe['Treatment'].iloc[i]="Phototherapy, laser therapy, systemic medications, lifestyle mods"
  elif(dataframe['Medical Condition'].iloc[i]=="DEPRESSION"):
    dataframe['Medication'].iloc[i]="SSRIs, SNRIs, tricyclic antidepressants"
    dataframe['Treatment'].iloc[i]="Psychotherapy, electroconvulsive therapy, medication management"
  elif(dataframe['Medical Condition'].iloc[i]=='MIGRAINES'):
    dataframe['Medication'].iloc[i]="Triptans, beta-blockers, anticonvulsants"
    dataframe['Treatment'].iloc[i]="Lifestyle modifications, biofeedback, nerve block injections"
  elif(dataframe['Medical Condition'].iloc[i]=='ADHD'):
    dataframe['Medication'].iloc[i]="Stimulants, non-stimulants, alpha-2 agonists"
    dataframe['Treatment'].iloc[i]="Behavioral therapy, parent training, educational interventions"
  elif(dataframe['Medical Condition'].iloc[i]=='HYPERTENSION'):
    dataframe['Medication'].iloc[i]="ACE inhibitors, diuretics, beta-blockers"
    dataframe['Treatment'].iloc[i]="Lifestyle modifications, weight management, stress reduction"
  elif(dataframe['Medical Condition'].iloc[i]=='CHRONIC BACK PAIN'):
    dataframe['Medication'].iloc[i]="NSAIDs, muscle relaxants, opioids"
    dataframe['Treatment'].iloc[i]="Physical therapy, chiropractic care, spinal injections, surgery"
  elif(dataframe['Medical Condition'].iloc[i]=='ANXIETY DISORDER'):
    dataframe['Medication'].iloc[i]="SSRIs, benzodiazepines, beta-blockers"
    dataframe['Treatment'].iloc[i]="Cognitive-behavioral therapy, relaxation techniques, medication"
  elif(dataframe['Medical Condition'].iloc[i]=='DIABETES'):
    dataframe['Medication'].iloc[i]="Insulin, oral hypoglycemic agents"
    dataframe['Treatment'].iloc[i]="Blood sugar monitoring, insulin therapy, dietary management"
```

```python
#Identifying symptoms from user prompt
con_symptom=[]
for start in range(0,len(prompt),1):
  for end in range(start,len(prompt)+1,1):
    if(prompt[start:end] in symptom_list):
      con_symptom.append(prompt[start:end])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
# Vectorize symptoms in the dataframe and user input
vectorizer = TfidfVectorizer()
symptoms_matrix = vectorizer.fit_transform(dataframe['Symptoms'].fillna(''))
user_input_vector = vectorizer.transform([" ".join(con_symptom)])
```

```python
# Calculate cosine similarity
cosine_similarities = linear_kernel(user_input_vector, symptoms_matrix).flatten()
# Get the index of the highest similarity
highest_similarity_index = cosine_similarities.argmax()
# Print the matched row
matched_row = dataframe.iloc[highest_similarity_index]
```
