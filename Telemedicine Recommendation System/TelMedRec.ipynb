```python
#pip install chardet
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd
import chardet
```

```python
def preprocess_text(input_text):
    cleaned_text = re.sub(r'[^\w\s]', '', input_text) #this removes all special characters and symbols from the prompt.
    cleaned_text = cleaned_text.lower() #converts the whole text to lower case.
    tokenized_words = word_tokenize(cleaned_text) #Tokenises the whole sentence
    stop_words = set(stopwords.words('english')) # This line retrieves a set of English stopwords.
    #Stopwords are common words that often do not carry significant meaning in text analysis, such as "and", "the", "is", etc.
    final_words = [word for word in tokenized_words if word not in stop_words]
    return final_words
```

```python
def find_matching_tokens(tokens, dataframe):
    matched_tokens = []
    for token in tokens:
        matching_rows = dataframe[dataframe['Symptoms'].str.contains(token, case=False)]
        if not matching_rows.empty:
            matched_tokens.append(token)
    return matched_tokens
```

```python
#Created New DateFrames
report_disease = pd.DataFrame(columns=['disease_name'])
add_dis= pd.DataFrame(columns=['Symptoms'])
```

```python
list_from_df=[]
```

```python
'''
This code iterates over each token in the processed_tokens list and then iterates over each row in DataFrame df2.
For each token, it checks if the token matches any value in the current row of df2. If there's a match, it extracts the
disease name from the first column of the row and adds it to the report_disease DataFrame.
'''
for i in  processed_tokens: #Here I choose symptom from user prompt to compare it with dataframe info.
    for index1, row in df2.iterrows(): #Index will be stored in index, row will store the rows of dataframe, as its iterrows
        list_from_df = row.tolist() #Converting series object row to list.
        for index in range(len(list_from_df)):
            if i.lower() == list_from_df[index].lower(): #Choosing symptoms from list of each row and comparing it to
                flag=list_from_df[0] #If true we store the disease name in flag.
                new_row = {'disease_name': flag} #creating a new row for the dataframe report disease.
                #Code below will add the symptoms of the matching diseases to add_dis dataframe.
                for sym in range( len(row[1:])) :
                    if (row[sym]!='null'):
                        new_row2 = {'Symptoms': row[sym]}
                        add_dis=add_dis.append(new_row2, ignore_index=True)
                report_disease = report_disease.append(new_row, ignore_index=True)
                break
```

```python
model = BioGptForCausalLM.from_pretrained("microsoft/biogpt") #Loads a pre-trained BioGPT model

tokenizer = BioGptTokenizer.from_pretrained("microsoft/biogpt")

generator = pipeline('text-generation', model=model, tokenizer=tokenizer)

set_seed(42) #Line sets the random seed to ensure that the generated text is reproducible.
```

```python
listOfDiseases=list(possibleDiseases['disease_name'])
```

```python
def prompt(Input1,Input2):
  lifestyleChanges.append(generator(Input1, max_length=100000000, num_return_sequences=1, do_sample=True))
  medicines.append(generator(Input2, max_length=100000000, num_return_sequences=1, do_sample=True))
```

```python
lifestyleChanges=[]
medicines=[]
for i in listOfDiseases:
  Input1="Lifestyle changes to be adopted for "+i+" are as follows"
  Input2="Medicines to be used for "+i+" are as follows"
  prompt(Input1,Input2)
```
