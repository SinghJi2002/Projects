# Flower Image Classifier

```python
#Removing warnings
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' # tensorflow INFO and WARNING messages are not printed 

#Importing important libraries
import pathlib
import numpy as np
import matplotlib.pyplot as plt
import PIL
from PIL import Image, ImageOps
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import cv2
from itertools import accumulate
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits, load_wine
import tensorflow as tf
from tensorflow.keras import datasets
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pickle
```

```python
#Importing dataset
#Here we have downloaded the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML311-Coursera/datasets/flower_photos.tgz"

#The dataset that we have downloaded in zipped(compressed) manner. Using the get_file method of keras we are unzipping the 
#files in the dataset into the main directory where we are working.
data_dir = keras.utils.get_file(origin=dataset_url,
                                fname='flower_photos',
                                untar=True)

#Using pathlib, we are storing the directory where are images are going.
data_dir = pathlib.Path(data_dir)
```

```python
#Image preprocessing
img_width = 150
img_height = 150
pics = list()
pics_arr = list()
p_class = list()

plt.figure(figsize=(20,5))
for idx, folder in enumerate(data_dir.glob('[!LICENSE]*')):
    cat = list(data_dir.glob(folder.name + '/*'))
    pic = PIL.Image.open(str(cat[0])).resize((img_width, img_height))
    pic_arr = np.array(pic)
    clss = folder.name
    
    #Plotting
    plt.subplot(1,5,idx+1)
    plt.imshow(pic)
    plt.title(clss)
    plt.axis('off')
    
    #Storing of images
    pics.append(pic)
    pics_arr.append(pic_arr)
    p_class.append(clss)
    
```

```python
#Few more important variables
batch_size=64
```

```python
#Next step is to create, training, validation and test sets. For this we are going to use using the `ImageDataGenerator` 
#and `flow_from_directory` functions from `keras.utils`.

#Training dataset.
train_generator=ImageDataGenerator(validation_split=0.2,
                              rescale=1/255.0,
                              width_shift_range=0.2,
                              height_shift_range=0.2,
                              horizontal_flip=True)
#The defination of each of these attributes of ImageGenerator is given below.

train_set=train_generator.flow_from_directory(directory=data_dir,
                                              seed=10,
                                              class_mode='sparse',
                                              batch_size=batch_size,
                                              shuffle=True,
                                              target_size=(img_height,img_width),
                                              subset='training')
##The defination of each of these attributes of flow_from_directory is given below
```

```python
#Validation set
val_gen = ImageDataGenerator(validation_split=0.2, 
                                rescale=1.0/255.0,
                                width_shift_range=0.2, 
                                height_shift_range=0.2,
                                horizontal_flip=True)
val_set = val_gen.flow_from_directory(
                               directory=data_dir,
                               seed=10,
                               class_mode='sparse',
                               batch_size=batch_size,
                               shuffle=True,
                               target_size=(img_height, img_width),
                               subset='validation')
```

```python
#List of class names of flowers
class_names = {y: x for x, y in val_set.class_indices.items()}
class_names
```

```python
#Now we will create our model
model=Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), padding='same', input_shape= (img_width, img_height, 3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same',activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2,2)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss="sparse_categorical_crossentropy",
              metrics=['accuracy'])
```

```python
model.fit(
  train_set,
  validation_data=val_set,
  epochs=40
)
```

```python
model.save('model.keras')
```

```python
#Predicting using this trained model
image=pics_arr[2]
plt.imshow(np.array(image, dtype='int'))
print(image.shape)
#Reshaping image to fit model input parameters
img_tensor = image.reshape(1,150,150,3)
print(img_tensor.shape)
```

```python
#Prediction section continued
y_pred=model.predict(img_tensor)
label = class_names[np.argmax(y_pred)]

#Ploting the result
plt.imshow(img_tensor.reshape((img_width,img_height,3)).astype("uint8"))
plt.title(f"Predicted class is: {label}", fontsize=13)
```
